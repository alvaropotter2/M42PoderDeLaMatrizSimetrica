```python
N = 100
A = np.zeros((N,N))

for i in np.arange(0,N):
    for j in np.arange(i,N):
        A[i][j] = np.random.randn(1)*(-1)**(i+j)
        A[j][i] = A[i][j]
    

A = A + np.diag(np.ones(N))
```

```python
plt.figure(figsize=(10,10))
plt.imshow(A)
plt.show()
```
